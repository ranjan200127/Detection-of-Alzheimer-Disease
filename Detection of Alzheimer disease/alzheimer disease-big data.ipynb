```python
import pandas as pd
```

```python
data = pd.read_csv('DWT_wavelets1.csv')
data.drop(data.columns[[0]], axis=1, inplace=True)
```

```python
data.head()
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('abc').getOrCreate()
sdf = spark.createDataFrame(data)
```

```python
features=[str(i) for i in range(1,2531)]
```

```python
from pyspark.ml.feature import VectorAssembler
```

```python
assembler = VectorAssembler(inputCols=features,outputCol="features")
```

```python
finalData=assembler.transform(sdf)
```

```python
finalData.show()
```

```python
finalized_data=finalData.select("features","2531")
```

```python
from pyspark.ml.tuning import ParamGridBuilder, TrainValidationSplit

training, test = finalized_data.randomSplit([0.7, 0.3], seed=8) 
```

```python
from pyspark.ml.classification import LogisticRegression
```

```python
lr = LogisticRegression(labelCol="2531", featuresCol="features",maxIter=10)
```

```python
model=lr.fit(training)
```

```python
predict_test=model.transform(test)
```

```python
predict_test.select("2531","prediction").show(10)
```

```python
import pyspark.sql.functions as F
check = predict_test.withColumn('correct', F.when(F.col('2531') == F.col('prediction'), 1).otherwise(0))
check.groupby("correct").count().show()
```
